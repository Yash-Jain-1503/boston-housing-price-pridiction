```python
import numpy as np
```

```python
x=data['dis'].values
y=data['medv'].values
```

```python
mean_x= np.mean(x)
mean_y= np.mean(y)
```

```python
n=len(x)
numer_a=0
numer_b=0
denom=0
```

```python
for i in range(n):
    numer_a += (x[i]*y[i])-(mean_x*mean_y)
    numer_b += (mean_y*(x[i]**2))-(mean_x*(x[i]*y[i]))
    denom +=  (x[i]**2)-(mean_x**2)
 
a= numer_a/denom
b= numer_b/denom

max_x = np.max(x)
min_x = np.min(x)
X = np.linspace(min_x,max_x,100)
Y= a*X + b

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] =[20.0,10.0]




plt.plot(X,Y,color='red')
plt.scatter(x,y,c='green')


plt.xlabel('dis from botson employment centers')
plt.ylabel('house price')
plt.show()

```
